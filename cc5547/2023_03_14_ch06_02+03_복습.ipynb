```python
#@title 합성

customers = ['Alice', 'Ben', 'Chales', 'Don', 'Ein']
ids = [f"1{i+1:03d}" for i in range(len(customers))]
ids, customers
```

```python
df1 = pd.DataFrame({
    "id": ids,
    "name": customers,
})
df1
```

```python
df2 = pd.DataFrame({
    "id": np.random.choice([f"1{i+1:03d}" for i in range(len(customers)+2)], 10, replace=True)
})
df2
```

```python
df2.info()
```

```python
df2['money'] = df2.id.apply(lambda x: int(x) * abs(np.random.randn()) * 1000).apply(int)
df2
```

```python
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='outer')
```
