````python
x = np.array([1, 2, 3], dtype='i')
x2 = np.array([1, 2, 3], dtype=int)
print(x, x.dtype)
print(x2, x2.dtype)
````

````python
x = np.array([1, 2, 3], dtype='f')
x2 = np.array([1, 2, 3], dtype=float)
print(x, x.dtype)
print(x2, x2.dtype)
````

````python
x = np.array([1, 2, 3.5], dtype='i')
print(x, x.dtype)
````

````python
x = np.array([1, 2, "안녕"], dtype='i')
print(x, x.dtype)
````
